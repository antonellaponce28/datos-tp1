### 49. Si decimos que un usuario sabe un idioma cuando tiene un nivel de babel mayor o igual a 1, para aquellos que editaron una de las versiones actuales del contenido, ¿Cuál es la tasa de revisiones sin comentario que realizan en función de los idiomas que saben? (⭐⭐⭐)

```python
languages_sabidos = languages.loc[languages['babel_level'] >= 1, :]
languages_sabidos.head()
```

```python
lang_subset = languages_sabidos.drop_duplicates(subset = 'babel_user')
lang_subset.head()
```

```python
contents['revisor_comment'].isnull().sum() #veo la cantidad de comentarios vacios, solo para saber
```

```python
contents_reduced = contents.loc[:,['revisor_id', 'revisor_comment']]
contents_reduced
```

```python
contents_reduced['revisor_id'].value_counts()
```

```python
comments = lang_subset.merge(contents_reduced, right_on = 'revisor_id', left_on = 'babel_user')\
    .loc[:,['babel_user', 'revisor_comment']]
comments.head()
```

```python
comments['comment_NaN'] = comments['revisor_comment'].isnull()
comments.head()
```

```python
comments_by_user = comments.groupby('babel_user').agg({'revisor_comment' : 'count', 'comment_NaN' : 'sum'})
comments_by_user = comments_by_user.reset_index()
comments_by_user
#Algunos usuarios tienen 0 revisiones y más de 0 revisiones sin comentarios por lo que los datos no son consistentes
#para esos casos, por lo tanto procedo a eliminar esos registros
#Además, no tiene sentido calcular una tasa para aquellos que no tienen revisiones
```

```python
comments_by_user.rename(columns = {'revisor_comment' : 'revisiones', 'comment_NaN' : 'revisiones_sin_comentario'}\
                        , inplace = True)
```

```python
comments_by_user = comments_by_user.loc[comments_by_user['revisiones'] != 0, :]
comments_by_user
```

```python
comments_by_user['tasa_revision_sin_comment'] =\
    comments_by_user['revisiones_sin_comentario'] / comments_by_user['revisiones']
comments_by_user
```

```python
comments_by_user.loc[:, ['babel_user', 'tasa_revision_sin_comment']]
```

### 50. Si decimos que un usuario sabe un idioma cuando tiene un nivel de babel mayor o igual a 1 consiga un dataframe cuyas columnas son tipos de logs, el índice es la cantidad de idiomas que sabe un usuario y las celdas la probabilidad de que esos usuarios generen ese tipo de log. (⭐⭐⭐)

```python
#languages_sabidos = languages.loc[languages['babel_level'] >= 1, :]
#languages_sabidos
```

```python
languages_sabidos['cantidad_idioma'] = 1
languages_sabidos
```

```python
lang_cantidad_idiomas = languages_sabidos.groupby('babel_user').agg({'cantidad_idioma' : 'sum'}).reset_index()
lang_cantidad_idiomas
```

```python
logs_logtype = logs.loc[:, ['contributor_id', 'logtype']]
logs_logtype
```

```python
lang_merge_logs = lang_cantidad_idiomas.merge(logs_logtype, right_on = 'contributor_id', left_on = 'babel_user')\
    .drop_duplicates()
```

```python
lang_merge_logs
```

```python
lang_logs_group = lang_merge_logs.groupby(['logtype', 'cantidad_idioma']).agg({'babel_user' : 'count'})
lang_logs_group
```

```python
lang_logs_group.reset_index()
```

```python
lang_logs_group
```

```python
lang_logs_pivot = lang_logs_group.reset_index()\
    .pivot(index = 'cantidad_idioma', columns = 'logtype', values = 'babel_user').fillna(0)
lang_logs_pivot
```

```python
suma_df = lang_logs_pivot.sum().sum()
suma_df
```

```python
lang_logs_probabilidades = lang_logs_pivot / suma_df
lang_logs_probabilidades
```

```python
#Chequeo que la suma de todo el df me de 1 (da casi 1)
lang_logs_probabilidades.sum().sum()
```
