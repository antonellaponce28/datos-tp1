## Dataset: Predicción de éxitos en oportunidades comerciales

```python
frio_frio = pd.read_csv("Entrenamieto_ECI_2020.csv")
```

```python
frio_frio.head()
```

```python
frio_frio_region = frio_frio.loc[frio_frio['Stage'] == 'Closed Won', :].groupby('Region').agg({'Stage' : 'count'})
frio_frio_region.reset_index(inplace = True)
frio_frio_region
```

```python
frio_frio.columns
```

```python
frio_frio["Value_ASP_converted"] = frio_frio["ASP_(converted)"] / frio_frio["ASP"]
frio_frio["Total_Amount_converted"] = frio_frio["Total_Amount"] * frio_frio["Value_ASP_converted"]
```

```python
frio_frio.info()
```

```python
frio_frio['Total_Amount_converted'] = frio_frio['Total_Amount_converted'].replace([np.inf, -np.inf], np.nan).dropna()
```

```python
frio_frio["Total_Amount_converted_ln"]= frio_frio["Total_Amount_converted"].apply(lambda x: np.log(x) if x!=0 else x)
```

### Visualización 3: Violin Plot

```python
sns.set(rc={'figure.figsize':(20,10)})

# Just switch x and y
ax = sns.violinplot( x=frio_frio["Stage"], y=frio_frio["Total_Amount_converted_ln"])


# Calculate number of obs per group & median to position labels
#medians = frio_frio.groupby(['Stage'])['ln Total Amount converted'].median().values
medians = frio_frio.groupby(['Stage'])['Total_Amount_converted_ln'].median().values
nobs = frio_frio['Stage'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]
 
# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
   ax.text(pos[tick], medians[tick] + 0.03, nobs[tick], horizontalalignment='center', color='black', weight='semibold')

#Titles
ax.set_title('Distribución de Total Amount [ln] por Stage',fontsize=20)
ax.set_xlabel('Stage',fontsize=15)
ax.set_ylabel('Total Amount',fontsize=15)
```

### Visualización 4: Bar Plot

```python
#plt.figure(figsize=(10,10))
 
# Choose the height of the bars
height = list(frio_frio_region["Stage"])
 
# Choose the names of the bars
bars = list(frio_frio_region["Region"])
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height, color= 'cornflowerblue' )
    
# Rotation of the bars names
plt.xticks(y_pos, bars, rotation=90)
 
# Custom the subplot layout
plt.subplots_adjust(bottom=0.4, top=0.99)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Oportunidades Closed Won',fontsize=15)
plt.title("Oportunidades Closed Won para cada Región",fontsize=20)
```

# Dataset: Partidos de Scaloni con la Selección Argentina
#### Fuente: https://www.transfermarkt.com.ar/lionel-scaloni/leistungsdatenDetail/trainer/62271

```python
scaloni = pd.read_csv("seleccion_scaloni.csv", sep = ';', encoding='latin-1')
```

```python
scaloni.head()
```

```python
resultados = scaloni['Resultado'].str.split(':', expand = True)
    #.rename(columns = {'0' : 'Resultado_Equipo_Local', '1' : 'Resultado_Equipo_Visitante'}, inplace = True)
resultados.head()
```

```python
resultados.columns = ['Resultado_Equipo_Local', 'Resultado_Equipo_Visitante']
resultados.head()
```

```python
scaloni.head()
```

```python
scaloni = pd.concat([scaloni, resultados], axis=1)
scaloni.head()
```

```python
scaloni['Resultado_Equipo_Local'] = pd.to_numeric(scaloni['Resultado_Equipo_Local'])
scaloni['Resultado_Equipo_Visitante'] = pd.to_numeric(scaloni['Resultado_Equipo_Visitante'])
```

```python
scaloni.info()
```

```python
scaloni['Ganados'] = ( ( scaloni['Equipo local'] == 'Argentina' ) \
    & ( scaloni['Resultado_Equipo_Local'] > scaloni['Resultado_Equipo_Visitante'] ) ) \
    | ( ( scaloni['Equipo visitante'] == 'Argentina' )\
      & ( scaloni['Resultado_Equipo_Visitante'] > scaloni['Resultado_Equipo_Local'] ) )
scaloni.head()
```

```python
scaloni['Perdidos'] = ( ( scaloni['Equipo local'] == 'Argentina' ) \
    & ( scaloni['Resultado_Equipo_Local'] < scaloni['Resultado_Equipo_Visitante'] ) ) \
    | ( ( scaloni['Equipo visitante'] == 'Argentina' )\
      & ( scaloni['Resultado_Equipo_Visitante'] < scaloni['Resultado_Equipo_Local'] ) )
scaloni.head()
```

```python
scaloni['Empatados'] = scaloni['Resultado_Equipo_Local'] == scaloni['Resultado_Equipo_Visitante']
scaloni.head()
```

```python
scaloni
```

```python
scaloni_partidos = scaloni.groupby('Competición').agg({'Ganados' : 'sum', 'Perdidos' : 'sum', 'Empatados' : 'sum' })
scaloni_partidos.reset_index(inplace = True)
scaloni_partidos
```

```python
plt.figure(figsize=(50,40))
# y-axis in bold
rc('font', weight='bold')
 
# Values of each group
bars1 = scaloni_partidos['Ganados']
bars2 = scaloni_partidos['Perdidos']
bars3 = scaloni_partidos['Empatados']
 
# Heights of bars1 + bars2
bars = np.add(0, bars3).tolist()
 
# The position of the bars on the x-axis
r = scaloni_partidos['Competición'].unique()
#r = [0,1,2,3,4,5]
 
# Names of group and bar width
#names = ['A','B','C','D','E']
barWidth = 1
 
# Create brown bars
plt.bar(r, bars1, color='#15B01A', edgecolor='white', width=barWidth, label = 'Ganados')
# Create green bars (middle), on top of the firs ones
plt.bar(r, bars2, color='#E50000',bottom=bars1, edgecolor='white', width=barWidth, label = 'Perdidos')
# Create green bars (top)
plt.bar(r, bars3, color='#FFD700',bottom=bars1+bars2, edgecolor='white', width=barWidth, label = 'Empatados')


# Custom X axis
plt.xticks(r, scaloni_partidos['Competición'].unique(), fontweight='bold', fontsize = 35)
plt.yticks(fontsize = 25)
plt.legend(loc="upper left", fontsize = 35)
plt.title("Desempeño de la Selección durante la ERA SCALONI", fontsize = 60, fontweight="bold")
plt.ylabel("Partidos", fontsize = 35)
# Show graphic
#plt.savefig("era_scaloni.jpg")
plt.show()
```
